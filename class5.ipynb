```python
# prompt: one question of each function and for loop

# Function Question:
# Write a function that takes a list of numbers and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
  """
  Calculates the sum of even numbers in a list.

  Args:
    numbers: A list of numbers.

  Returns:
    The sum of even numbers in the list.
  """
  sum_of_evens = 0
  for number in numbers:
    if number % 2 == 0:
      sum_of_evens += number
  return sum_of_evens

# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(f"The sum of even numbers is: {result}")


# For Loop Question:
# Write a for loop that iterates through a list of strings and prints each string in uppercase.

strings = ["hello", "world", "python"]

for string in strings:
  print(string.upper())
```
